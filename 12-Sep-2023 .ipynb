# RBG

## Red channel

```python
Red_image = np.copy(imaReadByCV)
print(Red_image.shape)
```

```python
Red_image[:,0]
```

```python
Red_image[:,:,1] = 0 # Green channel
```

```python
Red_image[:,:,2] = 0 # Blue channel 
```

```python
Red_image[:,0]
```

```python
plt.imshow(Red_image)
plt.show()
```

## Green channel

```python
Green_image = np.copy(imaReadByCV)
Green_image[:,:,0] = 0 # Red channel
Green_image[:,:,2] = 0 # Blue channel
plt.imshow(Green_image)
plt.show()
```

## Blue channel

```python
Blue_image = np.copy(imaReadByCV)
Blue_image[:,:,0] = 0 # Red channel
Blue_image[:,:,1] = 0 # Green channel
plt.imshow(Blue_image)
plt.show()
```

# Convert RGB to Gray scale

```python
gray_image = cv.cvtColor(imaReadByCV, cv.COLOR_BGR2GRAY)
```

```python
gray_image
```

```python
plt.imshow(gray_image)
```

# Convert RGB to HSV

```python
hsv_image = cv.cvtColor(imaReadByCV, cv.COLOR_BGR2HSV)
hsv_image
```

```python
plt.imshow(hsv_image)
plt.show()
```
