```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = cv2.imread('letter-i.jpg')
plt.imshow(img)
plt.show()
```

# Dilation Erosion

```python
kernel = np.ones((5,5),np.uint8)
```

```python
img_erosion = cv2.erode(img, kernel, iterations=1)
```

```python
img_dilation = cv2.dilate(img, kernel, iterations=1)
```

```python
cv2.imshow('Input', img)
cv2.imshow('Erosion', img_erosion)
cv2.imshow('Dilation', img_dilation)
cv2.waitKey(0)
```

```python
# Load  the image , convert it to grayscale
# and display it to our screen
image = cv2.imread('letter-i.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Original', image)
cv2.waitKey(0)
for i in range(0, 3):
    iterations = i + 1
    eroded = cv2.erode(gray.copy(), None, iterations=iterations)
    cv2.imshow(f"Eroded {iterations} times", eroded)
    cv2.waitKey(0)
```

```python
# close all windows to cleanup the screen
cv2.destroyAllWindows()

img = cv2.imread('letter-i.jpg')
cv2.imshow("Original", img)

for i in range(0, 3):
    iterations = i + 1
    dilated = cv2.dilate(gray.copy(), None, iterations=iterations)
    cv2.imshow(f"Dilated {iterations} times", dilated)
    cv2.waitKey(0)
```

# Opening

```python
# Close all windows to cleanup the screen, then initialize a list of 
# kernels sizes that will be applied to the image
cv2.destroyAllWindows()
cv2.imshow("Original", img)
kernelSizes = [(3, 3), (5, 5), (7, 7)]

# loop over the kernels sizes
for kernelSize in kernelSizes:
    # construct a rectangular kernel from the current size and then
    # apply an "opening" operation
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
    cv2.imshow(f"Opening: ({kernelSize[0]}, {kernelSize[1]})", opening)
    cv2.waitKey(0)
```
