# Tng cht lng hnh nh

```python
from PIL import ImageEnhance
```

## iu chnh sharpness

```python
enhancer = ImageEnhance.Sharpness(im)
enhancer.enhance(10.0)
```

## iu chnh contrast

```python
enhancer = ImageEnhance.Contrast(im)
enhancer.enhance(2)
```
