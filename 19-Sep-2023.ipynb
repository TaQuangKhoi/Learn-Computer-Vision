# Cắt hình

```python
im = Image.open('peacock.jpg')
box = (100, 150, 300, 300)
cropped_image = im.crop(box)
cropped_image
```

# Xoay hình

```python
rotated = im.rotate(180)
rotated
```

# Gộp 2 hình với nhau

```python
logo = Image.open('logo.png')
```

```python
# Gộp 2 hình theo vị trí xác định trước
position = (38, 469)
im.paste(logo, position)
```

```python
im.save('merged.jpg')
```

# Tăng chất lượng hình ảnh

```python
from PIL import ImageEnhance
```

## Điều chỉnh sharpness

```python
enhancer = ImageEnhance.Sharpness(im)
enhancer.enhance(10.0)
```

## Điều chỉnh contrast

```python
enhancer = ImageEnhance.Contrast(im)
enhancer.enhance(2)
```
