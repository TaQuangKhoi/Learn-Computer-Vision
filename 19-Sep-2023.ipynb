# Lm vic vi GIF

```python
from PIL import ImageSequence
```

```python
im = Image.open("embeddings.gif")
frame_num = 1
for frame in ImageSequence.Iterator(im):
    frame.save("frame%d.png" % frame_num)
    frame_num = frame_num + 1
    if frame_num == 3:
        break
```
